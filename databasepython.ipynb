```python
global Agenda,DNA,contactos
contactos = []
Agenda = {}
DNA = []

def MenuPrincipal(Agenda,contactos,DNA):

    print("1 - Introduzir um novo Cliente ")
    print("2 - Associar novos contactos para um Cliente ")
    print("3 - Associar um DNA de um Cliente ")
    print("4 - Testar gene de um DNA ")
    print("5 - Excluir um Cliente ")
    print("6 - Consultar Agenda ")
    print("7 - Gravar dados para um ficheiro ")
    print("8 - Ler dados de um ficheiro ")
    print("0 - Sair do programa")
    opcao =int(input("Escolha uma opção"))

    if opcao == 1:
        op1(Agenda,contactos,DNA)
    elif opcao ==2:
        op2(Agenda,contactos,DNA)
    elif opcao ==3:
        op3(Agenda, contactos,DNA)
    elif opcao ==4:
        op4(Agenda,contactos,DNA)
    elif opcao ==5:
        op5(Agenda,contactos,DNA)
    elif opcao ==6:
        op6(Agenda,contactos,DNA)
    elif opcao ==7:
        op7(Agenda,contactos,DNA)
    elif opcao ==8:
        op8()
    elif opcao ==0 :
        exit
    else:
        print("Erro.Introduza um número válido!")
        MenuPrincipal(Agenda,contactos,DNA)

        
        
#  Introdução de um novo cliente
def op1(Agenda,contactos,DNA):
    telemoveis=[]
    NomeCliente = input("Diga o nome do Cliente:")
    Agenda[NomeCliente]= telemoveis
    telemovel = int(input("Diga o contacto do Cliente (0 para terminar):"))
    telemoveis.append(telemovel)
    t = True
    while t:
        telemovel = int(input("Diga o contacto do Cliente (0 para terminar):"))

        if telemovel != 0 :
            telemoveis.append(telemovel)
            Agenda[NomeCliente] = telemoveis
        
        else:
            t = False
    contactos.append(telemoveis)
    print(Agenda)
    MenuPrincipal(Agenda,contactos,DNA)

    
    
#  Associar novos contactos para um cliente
def op2(Agenda,contactos,DNA):
    telemoveis=[]
    if Agenda == {}:
        print("Erro. Cliente não encontrado!")
        op1(Agenda,contactos,DNA)
    else:
        NomeCliente = input("Diga o nome do cliente para adicionar:")
        if Agenda[NomeCliente]:
            telemovel = int(input("Diga o contacto do Cliente (0 para terminar):"))
            telemoveis.append(telemovel)
            x = True
            while x:
                telemovel = int(input("Diga o contacto do Cliente (0 para terminar):"))

                if telemovel != 0:
                    telemoveis.append(telemovel)
                    Agenda[NomeCliente] = telemoveis
                    print ('Contacto adicionado!')
                else:
                    x = False
            contactos.append(telemovel)
            Agenda[NomeCliente]= telemoveis
    print(Agenda)
    MenuPrincipal(Agenda, contactos,DNA)


    
    
#    Adição de um gene
def op3(Agenda,contactos,DNA):
    contactos=[]
    DNA=[]
    aminoacidos=['A','T','G','C']
    if Agenda == {}:
        menu1(Agenda, contactos,DNA)
    else:
        NomeCliente = input("Diga o nome do cliente para adicionar")
        if NomeCliente in Agenda.keys():
            contactos=Agenda.get(NomeCliente)
            DNA = input("Insira a sequência de DNA: ").upper()
            DNA=[i for i in DNA]
            for i in range(len(DNA)):
                if DNA[i] in aminoacidos:
                    pass
                else:
                    print("ERRO: A sequência de DNA não é valida", "\n")
                    menu3(Agenda,contactos,DNA)

            DNA = "".join(DNA)
            contactos=contactos.append(DNA)
            print('Gene adicionado!')
           
        else:
            print("Contacto nao encontrado")
            MenuPrincipal(Agenda, contactos,DNA)

    print(Agenda)
    MenuPrincipal(Agenda,contactos,DNA)


    
    
#   Verificação da presença de um gene XPT
def op4(Agenda,contactos,DNA):
    if Agenda == {}:
        menu1(Agenda,contactos,DNA)
    else:
        NomeCliente = input("Diga o nome do cliente para teste do seu DNA")
        if NomeCliente in Agenda.keys():
            DNA= Agenda.get(NomeCliente)
            if DNA[0:3] == "ATG" and len(DNA) % 3 == 0:
                if DNA[-3:] == "TAG" or "TAA" or "TGA":
                    print("O gene é XPT")

            else:
                print("O gene nao é XPT")
    MenuPrincipal(Agenda,contactos,DNA)

    
    
#     Eliminação de um cliente
def op5(Agenda,contactos,DNA):
    NomeCliente = input("Diga o nome do cliente")
    if NomeCliente in Agenda.keys():
        Agenda.pop(NomeCliente)
    print(Agenda)
    MenuPrincipal(Agenda,contactos,DNA)

    




#     Print da Agenda no momento
def op6 (Agenda,contactos,DNA):
    NomeCliente = input("Diga o nome do cliente")
    if NomeCliente in Agenda.keys():
        print(Agenda)
        MenuPrincipal(Agenda,contactos,DNA)

        
        
#     Guardar um ficheiro formato txt
def op7(Agenda,contactos,DNA) :
    Agenda = Agenda
    f = open ("Agenda.txt","w")
    f.write (str(Agenda))
    f.close()
    MenuPrincipal(Agenda,contactos,DNA)

if __name__ == '__main__':
    MenuPrincipal(Agenda,contactos,DNA)






```
